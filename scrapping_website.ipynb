```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

BASE_URL = 'https://aplikasi-indonesia.com/gaul?page='

# Inisialisasi list untuk menyimpan data
data = []
# Inisialisasi set untuk menyimpan singkatan yang sudah diproses
processed_singkatan = set()

page_number = 68
while True:
    print(page_number)
    response = requests.get(BASE_URL + str(page_number))
    
    # Cek jika halaman kosong atau tidak ditemukan, berhenti looping
    if response.status_code != 200:
        break

    soup = BeautifulSoup(response.content, 'html.parser')

    # Mengambil semua link singkatan
    links = [a['href'].replace('gaul/', '') for a in soup.select('a[href^="gaul/"]')]

    for singkatan in links:
        # Hanya proses singkatan yang belum diproses sebelumnya
        if singkatan not in processed_singkatan:
            link = 'https://aplikasi-indonesia.com/gaul/' + singkatan
            response = requests.get(link)
            soup = BeautifulSoup(response.content, 'html.parser')

            element = soup.select_one('font[style="color: #ff8763;"]')
            
            if element:  # Cek jika elemen ditemukan
                kata = element.text
                # Menambahkan hasil scraping ke dalam list
                data.append({
                    'Singkatan': singkatan,
                    'Kata': kata
                })
                # Tambahkan singkatan ke set processed_singkatan
                processed_singkatan.add(singkatan)
    
    # Cek jika sudah mencapai halaman ke 2
    if page_number == 127:
        break

    page_number += 1  # Lanjut ke halaman berikutnya

# Membuat DataFrame dari list data
df = pd.DataFrame(data)

# Tampilkan DataFrame
print(df)
```

```python
df.to_csv('scraping_data_bahasa_gaul_68_127_page.csv',index = False)
```
